*Helper Function for the converting the dictionary into list which includes the only names of genres,keywords name,listing only top 3 actors of the movies .listing only the director name in crew column 
by using the ast.literal_eval()--- helping to avoid the string indices must be integer error *

```python
movies.dropna(axis=0,inplace=True)
```

```python
movies.info()
```

```python
import ast
def convert_dic_to_list(obj):
    l=[]
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l
    
```

```python
movies.iloc[0].genres
```

```python
movies['genres']=movies['genres'].apply(convert_dic_to_list)
```

```python
movies['genres']
```

```python
movies['keywords']=movies['keywords'].apply(convert_dic_to_list)
```

```python
movies['keywords']
```

```python
#we need to select top 3 actors from the cast 

def selecting_top_3_actors(obj):
    l=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter!=3:
            l.append(i['name'])
            counter+=1
        else:
            break
        
        
    return l
    
```

```python
movies['cast']=movies['cast'].apply(selecting_top_3_actors)
```

```python
movies['cast']
```

```python
#selecting the director name
movies['crew'].iloc[0]
```

```python
import ast
def movie_director(obj):
    name=[] 
    for i in ast.literal_eval(obj):
        if i.get('job')=='Director':
            name.append(i['name'])
    return name
    
```

```python
movies['crew']=movies['crew'].apply(movie_director)
```

```python
movies['crew']
```

#Observation:
**Successfully completed the selecting the  only names of genres,keywords name,listing only top 3 actors of the movies .listing only the director name in crew column**

```python
movies.head()
```

```python
#adding the each list to the every overview of the movie title
movies['overview']=movies['overview'].apply(lambda x:x.split())
movies['overview']
```

```python
movies.head()
```

```python
movies['genres']=movies['genres'].apply(lambda x:[i.replace(' ','')for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(' ','')for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(' ','')for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(' ','')for i in x])
```

```python
movies.head()
```

```python
#merging all the columns
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
movies.head()
```

```python
movies.columns.tolist()
```

```python
#now lets drop all the columns except movie_id,title,tags
movies=movies.drop(axis=1,columns=[ 'overview', 'genres', 'keywords', 'cast', 'crew'])
movies
```

```python
#now lets remove the list  in the tag and convert into string
movies['tags']=movies['tags'].apply(lambda x:' '.join(x))
```

```python
new_df=movies[['movie_id','title','tags']]
```

```python
new_df.head()
```

```python
#now lets convert all the string in the tags in each record. into lower case string
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

**Applying the Stemming Process using PortStemmer()**


```python
from nltk.stem import PorterStemmer
ps=PorterStemmer()
```

```python
#Lets Create the Helper Function
def stemming(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return ' '.join(y)
```

```python
new_df['tags'].iloc[0]
```

```python
new_df['tags']=new_df['tags'].apply(stemming)
new_df['tags']
```

**Text Vectorization using Sklearn Countvectorizer**

```python
#using machine learning library
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=5000)
vectors=cv.fit_transform(new_df['tags']).toarray()
vectors.shape
```

```python
vectors.shape
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
similarity.shape
```

```python
similarity[0]#Similarity check  for Avatar Movie with another movies
```

```python
def recommend_movies(movie):
    movie_index=new_df[new_df['title']==movie].index[0]
    distances=similarity[movie_index]#distances betweeen with different movies in the dataset (vector)
    movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]

    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend_movies('Avatar')
```

```python
recommend_movies('Batman Begins')
```

```python
recommend_movies('Epic')
```
